# T-Learner Uplift Model Training

This notebook visually walks through T-Learner uplift model training.

The uplift score is:

```text
uplift_score = P(conversion | treatment = 1, features) - P(conversion | treatment = 0, features)
```

This is the core modeling step of the project.
